```python
# 3D RECONSTRUCTION MODEL - Blocks World Scene Interpretation

import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import lil_matrix, csr_matrix
from scipy.sparse.linalg import lsqr
from mpl_toolkits.mplot3d import Axes3D

class BlocksWorld3DReconstructor:
    """3D reconstruction model for blocks world scenes"""
    
    def __init__(self, theta=30, weight_ground=1.0, weight_contact=1.0, 
                 weight_vertical=0.5, weight_horizontal=0.5, weight_smooth=0.1):
        """
        Initialize reconstruction model
        """
        self.theta = theta
        self.weights = {
            'ground': weight_ground,
            'contact': weight_contact,
            'vertical': weight_vertical,
            'horizontal': weight_horizontal,
            'smooth': weight_smooth
        }
    
    def pixel_to_index(self, y, x, width):
        """Convert 2D pixel coordinate to 1D vector index"""
        return y * width + x
    
    def build_constraint_system(self, img_shape, preprocessing_results):
        """
        Build sparse linear system Ax = b for recovering Z coordinates
        """
        h, w = img_shape
        n_pixels = h * w
        
        # Extract preprocessing results
        edges = preprocessing_results['edges']
        vertical_mask = preprocessing_results['vertical_mask']
        horizontal_mask = preprocessing_results['horizontal_mask']
        contact_edges = preprocessing_results['contact_edges']
        figure_mask = preprocessing_results['figure_mask']
        gx = preprocessing_results['gx']
        gy = preprocessing_results['gy']
        
        constraints = []
        constraint_idx = 0
        
        # Constraint 1: Ground pixels have Z = 0
        print("  Adding ground constraints...")
        for y in range(h):
            for x in range(w):
                if figure_mask[y, x] == 0:  # Ground pixel
                    idx = self.pixel_to_index(y, x, w)
                    constraints.append((constraint_idx, idx, 
                                      self.weights['ground'], 0.0))
                    constraint_idx += 1
        
        # Constraint 2: Contact edges have Z = 0
        print("  Adding contact edge constraints...")
        for y in range(h):
            for x in range(w):
                if contact_edges[y, x] > 0:
                    idx = self.pixel_to_index(y, x, w)
                    constraints.append((constraint_idx, idx, 
                                      self.weights['contact'], 0.0))
                    constraint_idx += 1
        
        # Constraint 3: Vertical edges - dZ/dy = 0
        print("  Adding vertical edge constraints...")
        edge_pixels = edges > 0
        for y in range(1, h-1):
            for x in range(w):
                if edge_pixels[y, x] and vertical_mask[y, x] and figure_mask[y, x] > 0:
                    idx_current = self.pixel_to_index(y, x, w)
                    idx_below = self.pixel_to_index(y+1, x, w)
                    # Z[y+1,x] - Z[y,x] = 0
                    constraints.append((constraint_idx, idx_below, 
                                      self.weights['vertical'], 0.0))
                    constraints.append((constraint_idx, idx_current, 
                                      -self.weights['vertical'], 0.0))
                    constraint_idx += 1
        
        # Constraint 4: Horizontal edges - derivative along edge = 0
        print("  Adding horizontal edge constraints...")
        for y in range(1, h-1):
            for x in range(1, w-1):
                if edge_pixels[y, x] and horizontal_mask[y, x] and figure_mask[y, x] > 0:
                    # Tangent direction (perpendicular to gradient)
                    tx = -gy[y, x]  # Perpendicular to gradient
                    ty = gx[y, x]
                    norm = np.sqrt(tx**2 + ty**2)
                    
                    if norm > 1e-5:
                        tx /= norm
                        ty /= norm
                        
                        # Approximate directional derivative
                        if abs(tx) > abs(ty):
                            idx_current = self.pixel_to_index(y, x, w)
                            idx_right = self.pixel_to_index(y, x+1, w)
                            constraints.append((constraint_idx, idx_right, 
                                              self.weights['horizontal'], 0.0))
                            constraints.append((constraint_idx, idx_current, 
                                              -self.weights['horizontal'], 0.0))
                            constraint_idx += 1
        
        # Constraint 5: Planarity (smoothness) - second derivatives = 0
        print("  Adding smoothness constraints...")
        for y in range(1, h-1):
            for x in range(1, w-1):
                if figure_mask[y, x] > 0 and edges[y, x] == 0:
                    idx_current = self.pixel_to_index(y, x, w)
                    idx_left = self.pixel_to_index(y, x-1, w)
                    idx_right = self.pixel_to_index(y, x+1, w)
                    idx_up = self.pixel_to_index(y-1, x, w)
                    idx_down = self.pixel_to_index(y+1, x, w)
                    
                    # dZ/dx = 0: Z[x-1] - 2*Z[x] + Z[x+1] = 0
                    constraints.append((constraint_idx, idx_left, 
                                      self.weights['smooth'], 0.0))
                    constraints.append((constraint_idx, idx_current, 
                                      -2*self.weights['smooth'], 0.0))
                    constraints.append((constraint_idx, idx_right, 
                                      self.weights['smooth'], 0.0))
                    constraint_idx += 1
                    
                    # dZ/dy = 0: Z[y-1] - 2*Z[y] + Z[y+1] = 0
                    constraints.append((constraint_idx, idx_up, 
                                      self.weights['smooth'], 0.0))
                    constraints.append((constraint_idx, idx_current, 
                                      -2*self.weights['smooth'], 0.0))
                    constraints.append((constraint_idx, idx_down, 
                                      self.weights['smooth'], 0.0))
                    constraint_idx += 1
        
        return constraints, constraint_idx, n_pixels
    
    def solve_linear_system(self, constraints, n_constraints, n_pixels):
        """
        Solve overdetermined linear system Ax = b using sparse least squares
        """
        print(f"\nBuilding sparse matrix: {n_constraints} equations, {n_pixels} variables")
        
        # Build sparse matrix A and vector b
        A = lil_matrix((n_constraints, n_pixels))
        b = np.zeros(n_constraints)
        
        for constraint_idx, pixel_idx, coeff, value in constraints:
            A[constraint_idx, pixel_idx] += coeff  # Use += to handle multiple coefficients
            b[constraint_idx] = value
        
        # Convert to CSR format for efficient solving
        A = csr_matrix(A)
        
        print("Solving least squares system...")
        # Solve using sparse least squares (LSQR algorithm)
        result = lsqr(A, b, atol=1e-8, btol=1e-8, iter_lim=1000)
        z_vec = result[0]
        istop = result[1]
        itn = result[2]
        r1norm = result[3]
        
        # Print convergence info
        convergence_msg = {
            0: "x=0 is exact solution",
            1: "Converged: ||Ax-b|| small enough",
            2: "Converged: ||A'(Ax-b)|| small enough",
            3: "Converged: condition number limit reached",
            4: "||Ax-b|| small but ||A'(Ax-b)|| not small",
            5: "||Ax-b|| not small but ||A'(Ax-b)|| small",
            6: "Condition number too large",
            7: "Max iterations reached"
        }
        print(f"  Status: {convergence_msg.get(istop, 'Unknown')}")
        print(f"  Iterations: {itn}")
        print(f"  Residual norm: {r1norm:.6f}")
        
        return z_vec
    
    def recover_3d_coordinates(self, img_shape, z_vec):
        """
        Recover X, Y, Z world coordinates from solved Z values
        
        Using parallel projection equations from Chapter 2:
            x_img = X
            y_img = Y*cos() + Z*sin()
        
        Therefore:
            X = x_img
            Y = (y_img - Z*sin()) / cos()
            Z = solved from constraints
        """
        h, w = img_shape
        
        # Reshape Z from vector to image
        Z = z_vec.reshape(h, w)
        
        # Initialize coordinate arrays
        X = np.zeros((h, w))
        Y = np.zeros((h, w))
        
        # Convert angle to radians
        theta_rad = np.radians(self.theta)
        cos_theta = np.cos(theta_rad)
        sin_theta = np.sin(theta_rad)
        
        # Recover X and Y coordinates
        for y in range(h):
            for x in range(w):
                X[y, x] = x
                # Recover Y from projection equation
                if cos_theta != 0:
                    Y[y, x] = (y - Z[y, x] * sin_theta) / cos_theta
                else:
                    Y[y, x] = 0
        
        return X, Y, Z
    
    def reconstruct(self, img_shape, preprocessing_results):
        """
        Complete 3D reconstruction pipeline
        """
        print("3D RECONSTRUCTION")
        
        # Build constraint system
        print("\nBuilding constraint system...")
        constraints, n_constraints, n_pixels = self.build_constraint_system(
            img_shape, preprocessing_results)
        
        print(f"\nTotal constraints: {n_constraints}")
        print(f"  Ground: {sum(1 for c in constraints if c[2] == self.weights['ground'])}")
        print(f"  Contact: {sum(1 for c in constraints if c[2] == self.weights['contact'])}")
        print(f"  Vertical edges: ~{sum(1 for c in constraints if c[2] == self.weights['vertical'])//2}")
        print(f"  Horizontal edges: ~{sum(1 for c in constraints if c[2] == self.weights['horizontal'])//2}")
        print(f"  Smoothness: ~{sum(1 for c in constraints if c[2] == self.weights['smooth'])//3}")
        
        # Solve for Z coordinates
        z_vec = self.solve_linear_system(constraints, n_constraints, n_pixels)
        
        # Recover full 3D coordinates
        print("\nRecovering X, Y, Z world coordinates...")
        X, Y, Z = self.recover_3d_coordinates(img_shape, z_vec)
        
        print("\nReconstruction complete!")
        print(f"  Z range: [{Z.min():.2f}, {Z.max():.2f}]")
        print(f"  Y range: [{Y.min():.2f}, {Y.max():.2f}]")
        print(f"  X range: [{X.min():.2f}, {X.max():.2f}]")
        
        return X, Y, Z
    
    def visualize_3d_coordinates(self, img, X, Y, Z, preprocessing_results, 
                                 save_path='3d_reconstruction.png'):
        """
        Visualize recovered 3D world coordinates
        """
        fig = plt.figure(figsize=(18, 12))
        
        # Create grid
        gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)
        
        # Row 1: Original, Edges, Edge Classification
        ax1 = fig.add_subplot(gs[0, 0])
        ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        ax1.set_title('Original Image', fontsize=12, fontweight='bold')
        ax1.axis('off')
        
        ax2 = fig.add_subplot(gs[0, 1])
        ax2.imshow(preprocessing_results['edges'], cmap='gray')
        ax2.set_title('Detected Edges', fontsize=12, fontweight='bold')
        ax2.axis('off')
        
        ax3 = fig.add_subplot(gs[0, 2])
        edge_viz = np.zeros((*img.shape[:2], 3), dtype=np.uint8)
        edge_mask = preprocessing_results['edges'] > 0
        edge_viz[preprocessing_results['vertical_mask'] & edge_mask] = [0, 255, 0]
        edge_viz[preprocessing_results['horizontal_mask'] & edge_mask] = [255, 0, 0]
        ax3.imshow(edge_viz)
        ax3.set_title('Edge Classification\n(Green=Vertical, Red=Horizontal)', 
                     fontsize=12, fontweight='bold')
        ax3.axis('off')
        
        # Row 2: X, Y, Z coordinates
        ax4 = fig.add_subplot(gs[1, 0])
        im4 = ax4.imshow(X, cmap='gray')
        ax4.set_title('X World Coordinate', fontsize=12, fontweight='bold')
        ax4.axis('off')
        plt.colorbar(im4, ax=ax4, fraction=0.046)
        
        ax5 = fig.add_subplot(gs[1, 1])
        im5 = ax5.imshow(Y, cmap='gray')
        ax5.set_title('Y World Coordinate', fontsize=12, fontweight='bold')
        ax5.axis('off')
        plt.colorbar(im5, ax=ax5, fraction=0.046)
        
        ax6 = fig.add_subplot(gs[1, 2])
        im6 = ax6.imshow(Z, cmap='jet')
        ax6.set_title('Z World Coordinate (Depth)\nBlue=Near, Red=Far', 
                     fontsize=12, fontweight='bold')
        ax6.axis('off')
        plt.colorbar(im6, ax=ax6, fraction=0.046)
        
        # Row 3: 3D visualization and statistics
        ax7 = fig.add_subplot(gs[2, :], projection='3d')
        
        # Downsample for 3D plot (for performance)
        step = max(1, img.shape[0] // 50)
        X_down = X[::step, ::step]
        Y_down = Y[::step, ::step]
        Z_down = Z[::step, ::step]
        
        ax7.plot_surface(X_down, Y_down, Z_down, cmap='viridis', alpha=0.8)
        ax7.set_xlabel('X')
        ax7.set_ylabel('Y')
        ax7.set_zlabel('Z (Depth)')
        ax7.set_title('3D Surface Reconstruction', fontsize=12, fontweight='bold')
        
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        plt.show()
        
        print(f"\n3D reconstruction visualization saved to '{save_path}'")


# Usage

if __name__ == "__main__":
    import cv2
    import os
    # Import preprocessing module
    try:
        from preprocessing import ImagePreprocessor
    except ImportError:
        print("Error: preprocessing.py not found. Make sure it's in the same directory.")
        exit(1)
    
    # Create preprocessor and reconstructor
    preprocessor = ImagePreprocessor(edge_threshold=30, vertical_tolerance=15)
    reconstructor = BlocksWorld3DReconstructor(theta=30)
    
    # Process images
    for i in range(1, 9):
        img_path = f"image{i}.png"
        if os.path.exists(img_path):
            print(f"PROCESSING: {img_path}")
            
            # Load image
            img = cv2.imread(img_path)
            h, w = img.shape[:2]
            
            # Preprocessing
            print("\nStep 1: Preprocessing...")
            preprocessing_results = preprocessor.process(img)
            
            # 3D Reconstruction
            print("\nStep 2: 3D Reconstruction...")
            X, Y, Z = reconstructor.reconstruct((h, w), preprocessing_results)
            
            # Visualization
            print("\nStep 3: Visualization...")
            reconstructor.visualize_3d_coordinates(
                img, X, Y, Z, preprocessing_results,
                save_path=f'reconstruction_image{i}.png'
            )
            
            print(f"COMPLETED: {img_path}")
        else:
            print(f"Skipping {img_path} (not found)")
    print("ALL RECONSTRUCTIONS COMPLETE")
```
